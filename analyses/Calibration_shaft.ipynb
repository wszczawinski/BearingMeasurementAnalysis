```python
# libraries
import os
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#crate paths for csv files
CURR_DIR = os.getcwd()
print(CURR_DIR)
data_path = os.path.join(CURR_DIR, 'datasets/Calibration_shaft.csv')
```

```python
shaft_measurments = pd.read_csv(data_path, sep=';', decimal=',')
shaft_measurments = shaft_measurments.astype(float)
```

```python
shaft_measurments.head(-10)
```

```python
shaft_measurments['Count'] = shaft_measurments['Count']%1000

shaft_measurments['Count'] = shaft_measurments['Count']*360/1000
```

```python
shaft_measurments.head()
```

```python
plt.axes(projection = 'polar')

plt.plot(shaft_measurments['Z measurement'], shaft_measurments['Count'])
plt.rcParams['figure.figsize'] = [30, 7]
plt.show()
```
